```python
hours = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))

if hours > 40:
    # Calculate regular pay for 40 hours plus 1.5x for extra hours
    overtime = hours - 40
    pay = (40 * rate) + (overtime * rate * 1.5)
else:
    # Standard calculation
    pay = hours * rate
```
