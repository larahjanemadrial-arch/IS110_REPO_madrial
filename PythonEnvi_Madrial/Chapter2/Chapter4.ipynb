```python
import pandas as pd
import numpy as np
```

```python
def compute_pay(hours, rate):

    if hours <= 40:
        pay = hours * rate
    else:
        regular_pay = 40 * rate
        overtime_hours = hours - 40
        overtime_rate = rate * 1.5
        overtime_pay = overtime_hours * overtime_rate
        pay = regular_pay + overtime_pay

    return pay
```

```python
hours = 45
rate = 10
pay = compute_pay(hours, rate)
print(f"Pay for {hours} hours at ${rate}/hour: {pay}")
```

```python
def computegrade(score):
    if 98 <= score <= 100:
        return "1.0"
    elif 94 <= score < 97:
        return "1.25"
    elif 90 <= score < 93:
        return "1.5"
    elif 86 <= score < 89:
        return "1.75"
    elif 82 <= score < 85:
        return "2.0"
    elif 78 <= score < 81:
        return "2.25"
    elif 75 <= score < 77:
        return "2.5"
    elif 0 <= score < 74:
        return "2.0"
    else:
        return "Invalid Score"

# Testing the function
test_score = float(input("Enter score: "))
print("Grade:", computegrade(test_score))
```
